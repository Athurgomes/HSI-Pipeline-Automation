```python
import os
import sys

# --- INÍCIO DO CÓDIGO DE CORREÇÃO DE CAMINHO ---

# Obtém o caminho absoluto do diretório onde o notebook está localizado
# Ex: '/home/usuario/seu_projeto/notebooks'
notebook_path = os.getcwd()
print(f"Diretório atual do notebook: {notebook_path}")

# Para chegar à raiz do projeto, subimos um nível de diretório
# Ex: de '/home/usuario/seu_projeto/notebooks' para '/home/usuario/seu_projeto'
project_root = os.path.dirname(notebook_path)
print(f"Raiz do projeto inferida: {project_root}")

# Adiciona a raiz do projeto à lista de caminhos onde o Python procura por módulos
# Isso garante que ele encontrará a pasta 'src'
if project_root not in sys.path:
    sys.path.append(project_root)
    print(f"O caminho '{project_root}' foi adicionado ao sys.path.")

# --- FIM DO CÓDIGO DE CORREÇÃO DE CAMINHO ---


# Agora a importação original deve funcionar sem erros
try:
    from src.plotting import plotar_imagem_rgb, plotar_media_espectral
    print("\nSUCESSO: Funções locais importadas corretamente!")
except ImportError as e:
    print(f"\nERRO: A importação ainda falhou. Verifique a estrutura de pastas. Erro: {e}")

# O resto das importações padrão
import spectral
print("Bibliotecas padrão carregadas.")
```

```python
# Definindo o caminho para o arquivo de dados
# O '..' significa "subir um nível de pasta", saindo de 'notebooks' para a raiz do projeto.
caminho_base = '..'
arquivo_hdr = 'data/bronze level/hsi_original/ATCC27_240506-161129/capture/ATCC27_240506-161129.hdr'

# os.path.join cria o caminho de forma segura para qualquer sistema operacional
hdr_path_completo = os.path.join(caminho_base, arquivo_hdr)

# Carregando a imagem
try:
    img = spectral.open_image(hdr_path_completo)
    print(f"Imagem carregada com sucesso de: {hdr_path_completo}")
    print(f"Dimensões: {img.shape}")
except FileNotFoundError:
    print(f"ERRO: Arquivo não encontrado. Verifique o caminho: {hdr_path_completo}")
```
